### <a name='a2'></a> Załadowanie danych

````python
def fetch_financial_data(company='AMZN'):
    """Pobiera dane dotyczące notowań spółek na giełdzie według symboli
    na serwisie stooq.pl"""
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df = fetch_financial_data()
df.info()
````

### <a name='a6'></a> Obliczanie nowych kolumn

````python
df = fetch_financial_data('PLTR')
df.info()
````

````python
df.head()
````

````python
df.describe()
````

````python
df.head(3)
````

````python
df['Average'] = (df['Open'] + df['Close']) / 2
df.head()
````

````python
df = df.sort_index()
df.head()
````

````python
df['Close_shift'] = df[['Close']].shift(1)
df.head()
````

````python
df['Daily_Change'] = df.Close / df.Close_shift - 1
df.head()
````

````python
#del df['Daily Change']
````
