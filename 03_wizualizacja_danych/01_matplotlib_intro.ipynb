````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
````

````python
fig = plt.figure()
fig.show()
````

### <a name='1'></a> Wykres liniowy

````python
_ = plt.plot([0, 1], [0, 3])
````

````python
plt.plot([0, 1], [0, 3], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
````

````python
plt.plot([0, 1, 2], [0, 3, 2], label='line', color='green')
plt.xlabel('os x')
plt.ylabel('os y')
plt.title('Wykres')
_ = plt.legend()
````

````python
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
````

### <a name='2'> </a> Tworzenie wielu wykresów

````python
plt.subplot(121)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(122)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
````

````python
fig = plt.figure(figsize=(12, 8))

plt.subplot(131)
plt.plot([0, 1, 2], [3, 2, 5])
plt.subplot(132)
plt.plot([0, 1, 2], [3, 1, 4], 'r--')
plt.subplot(133)
plt.plot([0, 1, 2], [3, 1, 4], 'go')
````

### <a name='3'> </a> Wykresy słupkowe

````python
plt.bar(x=[0, 1, 2, 3], height=[1.2, 3.2, 1.2, 5.2])
````

````python
plt.bar(x=['a', 'b', 'c', 'd'], height=[1.2, 3.2, 1.2, 5.2])
````

````python
plt.barh(y=['a', 'b', 'c', 'd'], width=[1.2, 3.2, 1.2, 5.2])
````

### <a name='4'> </a> Wykresy punktowe

````python
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x=x1, y=x2)
````

````python
x3 = np.random.randn(300) * 50

plt.scatter(x=x1, y=x2, s=x3)
````

````python
plt.scatter(x=x1, y=x2, s=x3, alpha=0.5)
````
